```python
df = pd.read_csv('flights.txt',sep='\s+',header = None)
df
```

```python
#Avoiding flights with next day arrival
df2 = df[df.iloc[:,2] <= df.iloc[:,3]]
df2
```

```python
#Avoiding routes via other airports other than specified airports
allowedAirports = ['LAX', 'SFO', 'PHX', 'SEA', 'DEN', 'ATL', 'ORD', 'BOS', 'IAD', 'JFK']
df3 = df2[df2[0].isin(allowedAirports) & df2[1].isin(allowedAirports)]
df3
# pd.DataFrame(df3).to_csv('flights.csv',index=False)
```
